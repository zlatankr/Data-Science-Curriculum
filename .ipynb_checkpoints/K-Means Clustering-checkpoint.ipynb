# K-Means Clustering Algorithm

## Model Summary

K-means clustering is an unsupervised clustering algorithm used to find groups of similar objects. It belongs to the category of prototype-based clustering, which dictates that each cluster be represented by a prototype of either a _centroid_ (average), or a _medoid_ (the most representative or most frequently occurring point) in the case of categorical features. The algorithm also belongs to the broader category of vector quantization.

### How it works: 

1. Randomly pick _k_ centroids from the sample points as initial cluster centers. 
2. Assign each sample to the nearest centroid. 
3. Move the centroids to the center of the samples that were assigned to it.
4. Repeat the steps 2 and 3 until the cluster assignment do not change or a user-defined tolerance or a maximum number of iterations is reached.

The goal is to minimize the within-cluster sum of squared errors (SSE), which is calculated using the squared Euclidian distance formula between two points x and y in an m-dimensional space:  
![alt text](extras/EuclideanDistance.PNG)  
We sometimes refer to the cluster SSE as the cluster inertia. (In the case of k-means, the 'y' above is the centroid mean).

There are several assumptions that must be true for k-means to work:

- the variances of the distribution of each attribute is spherical
- all attributes have the same variance
- each cluster tends to be the same size (in terms of area, not in terms of cardinality).
- the data scientist must know the number of clusters (k) in the data


## Pros & Cons

Pros:
- Computationally-cheap algorithm
- Can be used for unsupervised learning (clustering)

Cons:
- One must know the number of clusters ahead of time.
- k-means will not fail when used on non-clustered data, but rather will just create clusters.
- Clusters must be spherical - this is easy to check in two-dimensions but much more difficult in high-N dimensions.
- k-means is strongly impacted by the "curse of dimensionality" due to its use of Euclidean distances. A dataset with many variables can perform poorly.
- It doesn't handle outliers well.
- Even with many iterations, the algorithm can converge on local minima.

## Coding Example

Let's use the iris dataset to see how an example of k-means clustering.

```python
import numpy as np
import pandas as pd
from sklearn import datasets
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = datasets.load_iris()
y = pd.DataFrame(iris['target'], columns=['target'])
X = pd.DataFrame(iris['data'], columns=['sepal length', 'sepal width', 'petal length', 'petal width'])
iris = pd.concat([X, y], axis=1)
```

What we want k-means to do is to identify the our three plant types according to sepal width and sepal length. 

```python
sns.lmplot(x='sepal length', y='sepal width', data=iris, hue='target', fit_reg=False)
```

Prepare data for modeling:

```python
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features. We could
                      # avoid this ugly slicing by using a two-dim dataset
y = iris.target
```

Implement the model:

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3,
    init='random',
    n_init=10,
    max_iter=300,
    tol=1e-04,
    random_state=0)
y_km = km.fit_predict(X)
```

Looking at the code above, we set our number of clusters to n_clusters=3. We set n_init=10 to run the k-means clustering algorithms 10 times independently with different random centroids to choose the final model as the one with the lowest SSE. Via the max_iter=300 parameter, we specify the maximum number of iterations for each single run.  

Note that the k-means implementation in scikit-learn stops early if it converges before the maximum number of iterations is reached. However, it is possible that k-means does not reach convergence for a particular run, which can be problematic (computationally expensive) if we choose relatively large values for max_iter. One way to deal with convergence problems is to choose larger values for tol, which is a parameter that controls the tolerance with regard to the changes in the within-cluster sum-squared-error to declare convergence. In the preceding code, we chose a tolerance of 1e-04 (=0.0001).  

Let's visually inspect our finished product: 

```python
plt.scatter(X[y_km == 0, 0],
            X[y_km == 0, 1],
            s=50,
            c='lightgreen',
            marker='s',
            label='cluster 1')
plt.scatter(X[y_km == 1, 0],
            X[y_km == 1, 1],
            s=50,
            c='orange',
            marker='o',
            label='cluster 2')
plt.scatter(X[y_km == 2, 0],
            X[y_km == 2, 1],
            s=50,
            c='lightblue',
            marker='v',
            label='cluster 3')
plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            s=250,
            marker='*',
            c='red',
            label='centroids')
plt.legend()
plt.grid()
plt.tight_layout()
#plt.savefig('./figures/centroids.png', dpi=300)
plt.show()
```

In this particular case, we can use the confusion matrix and the classification report to see how well the clusters predicted the actual targets. 

```python
from sklearn.metrics import classification_report, confusion_matrix
rep = {1:0, 0:1}
y_km2 = np.array(pd.Series(y_km).replace(rep))
print confusion_matrix(y, y_km2)
print '\n', classification_report(y, y_km2)
```

### K-means ++

As mentioned before, k-means initiates with a random placement of the k centroids on the feature space. Depending on how poorly the centroids are placed, this can result in bad clustering or slow convergence. We try to mitigate this by running the model algorithm multiple times (using the n_init parameter) and choosing the one with the lowest cluster inertia. However, another method we can use is k-means++, which initializes the centroids far away from each other, leading to better clusters and faster convergence. To use this method, we set the `init` parameter to `k-means++`.

### Finding the Optimal Number of Clusters: Elbow Method

In order to help us find the optimal number of clusters, we can compare the performance of different k-means clustersing models using our key metric of within-cluster SSE (known as distortion). Intuitively, we know that if k increases, distortion will decrease, since the samples will be closer to the centroids they are assigned to. The main idea behind the elbow method is to plot the distortion for different numbers of clusters and see at which k the distortion decreased the most drastically.  

Let's take a look:

```python
distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, 
                init='k-means++', 
                n_init=10, 
                max_iter=300, 
                random_state=0)
    km.fit(X)
    distortions.append(km.inertia_)
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
#plt.savefig('./figures/elbow.png', dpi=300)
plt.show()
```

Here, we can actually see that the elbow (the point at which the slope of the curve changes most rapidly) occurs at 2 clusters, which, although not the correct choice for the dataset, is actully makes a lot of sense when we look at the original chart:

```python
plt.scatter(X[:,0], X[:,1])
```

### Finding the Optimal Number of Clusters: Silhouette Plots

Silhouette analysis, which can be applied to algorithms other than k-means, is visual representation of how tightly grouped the same samples are in the clusters. To calculate the silhouette coefficient of a single sample in our dataset, we can apply the following three steps:
1. Calculate the cluster cohesion **a** as the average distance between a sample z and all other points in the same cluster.
2. Calculate the cluster separation **b** from the next closest cluster as the average distance between the sample x and all samples in the nearest cluster.
3. Calculate the silhouette **s** as the difference between cluster cohesion and separation divided by the greater of the two, as shown here:
![](extras/Silhouette.PNG)  

The silhouette coefficient is bounded in the range -1 to 1. Based on the preceding formula, we can see that the silhouette coefficient is 0 if the cluster separation and cohesion are equal **b**=**a**. Furthermore, we get close to an ideal silhouette coefficient of 1 if **b**>>**a**, since **b** quantifies how dissimilar a sample is to other clusters, and **a** tells us how similar it is to the other samples in its own cluster, respectively.  

The silhouette coefficient is available as `silhouette_samples` from scikit-learn's metric module, and optionally the `silhouette_scores` can be imported. This calculates the average silhouette coefficient across all samples, which is equivalent to `numpy.mean(silhouette_samples())`.  

By executing the following code, we will now create a plot of the silhouette coefficients for a k-means clustering with 3k=:

```python
import numpy as np
from matplotlib import cm
from sklearn.metrics import silhouette_samples

km = KMeans(n_clusters=2, 
            init='k-means++', 
            n_init=10, 
            max_iter=300,
            tol=1e-04,
            random_state=0)
y_km = km.fit_predict(X)

cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]
silhouette_vals = silhouette_samples(X, y_km, metric='euclidean')
y_ax_lower, y_ax_upper = 0, 0
yticks = []
for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    color = cm.jet(float(i) / n_clusters)
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0, 
             edgecolor='none', color=color)

    yticks.append((y_ax_lower + y_ax_upper) / 2.)
    y_ax_lower += len(c_silhouette_vals)
    
silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg, color="red", linestyle="--") 

plt.yticks(yticks, cluster_labels + 1)
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')

plt.tight_layout()
# plt.savefig('./figures/silhouette.png', dpi=300)
plt.show()
```
